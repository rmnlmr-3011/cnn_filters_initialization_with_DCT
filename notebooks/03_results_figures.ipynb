```python
import json
from pathlib import Path

import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from src.analysis.filter_analysis import analyze_model
from src.analysis.checkpoint_analysis import CUSTOM_OBJECTS

# ============================================================
# CONFIG
# ============================================================

runs_dir = Path("../runs")
output_dir = Path("../comparison/dct_component_energy_barplots")
output_dir.mkdir(parents=True, exist_ok=True)

selected_inits = ["he", "sigma", "sigma_pure", "dctlow", "dctlow_noise"]
show_figures = True

# Ordre voulu d'après ton schéma "Après"
# Grille voulue :
# 1 2 5
# 3 4 7
# 6 8 9
#
# Colonnes source actuelles (0-based) :
# 0 1 2
# 3 4 5
# 6 7 8
#
# Donc permutation 0-based :
display_order = [0, 1, 3, 4, 2, 6, 5, 7, 8]


# ============================================================
# HELPERS
# ============================================================

def build_run_index(runs_dir: Path) -> dict:
    runs_dir = Path(runs_dir)
    run_index = {}

    for run_dir in sorted(runs_dir.iterdir()):
        if not run_dir.is_dir():
            continue

        config_path = run_dir / "config.json"
        if not config_path.exists():
            continue

        with open(config_path, "r", encoding="utf-8") as f:
            config = json.load(f)

        init_name = config["initializer"]["name"]
        seed = int(config["seed"])
        run_index[(init_name, seed)] = run_dir

    return run_index


def select_best_seed_per_init(df_final: pd.DataFrame, selected_inits: list[str]) -> pd.DataFrame:
    df_sel = df_final[df_final["init"].isin(selected_inits)].copy()

    if df_sel.empty:
        raise ValueError("Aucune ligne trouvée dans df_final pour les initialisations demandées.")

    df_best = (
        df_sel
        .sort_values(["init", "test_accuracy"], ascending=[True, False])
        .groupby("init", as_index=False)
        .head(1)
        .reset_index(drop=True)
    )

    return df_best


def compute_mean_dct_component_energy(kernel_df: pd.DataFrame, display_order: list[int]) -> np.ndarray:
    """
    Retourne un vecteur de taille 9 :
    énergie moyenne de chaque composante DCT sur tous les noyaux,
    réordonnée selon l'ordre d'affichage voulu.
    """
    dct_energy_cols = [f"dct_energy_{i}" for i in range(9)]

    missing = [c for c in dct_energy_cols if c not in kernel_df.columns]
    if missing:
        raise ValueError(f"Colonnes manquantes dans kernel_df: {missing}")

    mean_energy = kernel_df[dct_energy_cols].mean(axis=0).to_numpy(dtype=float)

    total = mean_energy.sum()
    if total <= 0:
        raise ValueError("Somme des énergies moyennes <= 0, impossible de normaliser.")

    mean_energy = mean_energy / total

    # Réordonnancement selon ton schéma "Après"
    mean_energy = mean_energy[display_order]

    return mean_energy


def plot_dct_component_energy_barplot(
    mean_energy: np.ndarray,
    model_name: str,
    seed: int,
    save_dir: Path,
    show: bool = True,
):
    if mean_energy.shape != (9,):
        raise ValueError(f"mean_energy doit avoir shape (9,), reçu {mean_energy.shape}")

    cumulative = np.cumsum(mean_energy)
    x = np.arange(9)

    fig, ax = plt.subplots(figsize=(7, 5))

    ax.bar(x, mean_energy, width=0.8)

    ax.plot(
        x,
        cumulative,
        marker="x",
        linestyle=":",
        color="red",
        linewidth=2,
        markersize=8,
        label="Cumulative Energy",
    )

    ax.set_xlabel("DCT Component")
    ax.set_ylabel("Avg. Energy %")
    ax.set_title(f"{model_name} (best seed={seed})")
    ax.set_xticks(x)
    ax.set_xticklabels([str(i) for i in range(1, 10)])
    ax.set_ylim(0, 1.05)
    ax.legend()

    save_dir = Path(save_dir)
    save_dir.mkdir(parents=True, exist_ok=True)
    save_path = save_dir / f"{model_name}_seed{seed}_dct_component_energy.png"
    fig.savefig(save_path, dpi=150, bbox_inches="tight")

    if show:
        plt.show()
    else:
        plt.close(fig)

    return save_path


# ============================================================
# PIPELINE
# ============================================================

df_best = select_best_seed_per_init(df_final, selected_inits)
display(df_best[["init", "seed", "test_accuracy", "test_loss"]])

run_index = build_run_index(runs_dir)

saved_paths = []

for _, row in df_best.iterrows():
    init_name = row["init"]
    seed = int(row["seed"])

    key = (init_name, seed)
    if key not in run_index:
        raise KeyError(f"Run introuvable pour init={init_name}, seed={seed}")

    run_dir = run_index[key]
    model_path = run_dir / "model.keras"

    if not model_path.exists():
        raise FileNotFoundError(f"model.keras introuvable dans {run_dir}")

    print(f"[INFO] Loading model: init={init_name}, seed={seed}, path={model_path}")

    model = tf.keras.models.load_model(model_path, custom_objects=CUSTOM_OBJECTS)
    kernel_df, filter_df, layer_df, summary = analyze_model(model)

    mean_energy = compute_mean_dct_component_energy(
        kernel_df=kernel_df,
        display_order=display_order,
    )

    save_path = plot_dct_component_energy_barplot(
        mean_energy=mean_energy,
        model_name=init_name,
        seed=seed,
        save_dir=output_dir,
        show=show_figures,
    )

    saved_paths.append(save_path)

print("\nSaved figures:")
for p in saved_paths:
    print(f" - {p}")
```
